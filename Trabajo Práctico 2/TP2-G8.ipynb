## Ejercicios Teóricos

1. ¿Qué diferencia hay entre una estrategia de búsqueda Informada y una estrategia de búsqueda No Informada?

        Las estrategias de búsqueda no informada no utilizan ninguna información adicional sobre el problema más allá de lo que está explícitamente definido en el grafo de búsqueda (nodos, aristas).
        Las estrategias de búsqueda informada utilizan una función heurística para estimar la "bondad" o "prometedoridad" de los nodos, guiando la búsqueda hacia las soluciones más probables de manera más eficiente.

        Conclusión:

        Búsqueda No Informada: Sin conocimiento extra; explora sistemáticamente todos los posibles caminos.
        Búsqueda Informada: Utiliza una función heurística para dirigir la búsqueda hacia soluciones probables, haciendo el proceso más eficiente.


2. ¿Qué es una heurística y para qué sirve?

        Una heurística es una técnica o regla general que se utiliza para resolver un problema de manera rápida y eficiente, aunque no garantiza encontrar la solución óptima.

4. ¿Es posible que un algoritmo de búsqueda no tenga solución?

        Sí, es posible que un algoritmo de búsqueda no encuentre una solución si: El espacio de búsqueda no contiene una solución, es decir, no hay ningún camino que conecte el estado inicial con el estado objetivo. 
        El algoritmo no puede explorar completamente el espacio de búsqueda, como ocurre en algoritmos que no son completos (por ejemplo, la búsqueda en profundidad puede quedar atrapada en un bucle en espacios infinitos).
