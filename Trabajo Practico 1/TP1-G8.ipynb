**1. Defina con sus propias palabras inteligencia natural, inteligencia artificial y agente.**
    <br>La inteligencia es la capacidad cognitiva que poseen los seres humanos y algunos animales, la cual les permite aprender, razonar, resolver problemas y adaptarse a su entorno.<br>
    <br>Inteligencia Artificial: Es la disciplina de la informática que se enfoca en crear sistemas que simulan la inteligencia humana. Estos sistemas pueden aprender de experiencias, tomar decisiones y realizar tareas que requieren una inteligencia similar a la humana.
<br>
    <br>Un agente es cualquier entidad capaz de percibir su entorno a través de sensores y actuar sobre él utilizando actuadores.<br>

**2. ¿Qué es un agente racional?**

<br> Un agente es cualquier entidad capaz de percibir su entorno a través de sensores y actuar sobre él utilizando actuadores. Su racionalidad se define por su capacidad para cumplir con un criterio de rendimiento específico, que guía sus acciones para maximizar la eficiencia y efectividad en la consecución de sus objetivos.<br>

**5. Defina cada tipo de agente en función de su **estructura** y dé un ejemplo de cada categoría.
<br>Agente reactivo simple: Unicamente en base a una percepcion, tiene una salida. Algo continuo/lineal, 1 entrada se relaciona con una accion de salida (Algo directo que no se "analiza")<br>

<br>Agente reactivo complejo o con memoria: Tiene un objetivo al que yo quiero llegar y resuelve el problema de visibilidad parcial: almacena aquellas partes del mundo que no se ven, por lo que requiere de un estado interno que dependa de la historia de las percepciones.<br>
<br>Un Agente Basado en Utilidad es un tipo de agente que busca maximizar una función de utilidad, la cual evalúa el valor de diferentes resultados posibles. Este enfoque permite al agente tomar decisiones que no solo son racionales, sino que también optimizan el beneficio esperado.

<br>Agente reactivo basado en utilidad: Se centra mas en llegar la objetivo sin razonar tanto, solo cumplirlo

<br>Agentes que aprenden: compara la información del entorno, la compara con información previa y toma la decisión más adecuada.<br/>
