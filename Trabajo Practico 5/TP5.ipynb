1. ¿Cuáles son los tres axiomas de Kolmogorov?
los axiomas de Kolmogrov son:

        1- Todas las probalidades estan entre 0 y 1, para cualquier proposición a:
                        0 ≤ 𝑷(𝒂) ≤ 1

        2- Las proposiciones necesariamente ciertas tienen probabilidad 1 y las proposiciones necesariamente falsas tienen probabilidad 0
                P(cierto)=1  y  P(falso)=0

        3- La probabilidad de una diyunción viene dada por:
                P(a ∨ b) = P(a) + P(b) - P(a ∧ b)
