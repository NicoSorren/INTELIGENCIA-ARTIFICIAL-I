2. Describa las distintas heurísticas que se emplean en un problema de Satisfacción de Restricciones.

**MÍNIMOS VALORES RESTANTES**
Se centra en elegir la variable que tiene el menor número de valores posibles en su dominio. Esta estrategia prioriza las variables que están más cerca de quedarse sin opciones, con el objetivo de resolver primero las variables más restrictivas.

**GRADO HEURÍSTICO**
Intenta reducir el factor de ramificación sobre futuras opciones seleccionando la variable, entre las variables no asignadas, que esté implicada en el mayor número de restricciones. Al elegir una variable con muchas restricciones, se busca resolver primero los conflictos que afectan a más variables, lo que facilita la resolución del problema al reducir el impacto de las restricciones en el futuro.
La heurística del mínimo valores restantes es por lo general una guía más poderosa, pero el grado heurístico puede ser útil como un desempate

**VALOR MENOS RESTRINGIDO**
Se utiliza después de seleccionar una variable y consiste en elegir el valor que impone menos restricciones a las variables restantes. Esto se hace para mantener abiertas la mayor cantidad de opciones posibles para futuras asignaciones, minimizando el impacto de la elección en el resto del problema. En general, la heurística trata de dejar la flexibilidad máxima de las asignaciones de las variables siguientes.
