    El boleto es razonable de comprar cuando el valor esperado es mayor o igual al costo del boleto. En este caso, el boleto cuesta $1.

        Para el Juego B, el valor esperado es $0.50, que también es menor que el costo del boleto ($1). Por lo tanto, tampoco es razonable comprar un boleto para este juego basándonos en el valor esperado.

5. Una empresa minera quiere enviar a un robot recolector de materiales a visitar un planeta. El costo del viaje asciende a \$20.000.000. Tiene que decidir a qué planeta lo enviará de entre 5 opciones, sabiendo que uno de ellos contiene oro por un valor de \$100.000.000 y los no poseen nada de valor.<br>
La Comisión Nacional del Espacio ofrece los resultados de un estudio sobre el planeta X en el que indica si hay o no oro; y vende dicha información por \$25.000.000. ¿Nos beneficia comprar dicha información?

    C: $100M
    S: $20M
    n: 5 planetas
    I: $25M

    Comparamos el valor esperado cuando no tenemos la informacion frente al esperado cuanso si tenemos la informacion

    VE(~I): (1/5.C)+(4/5.0)-S=0
    VE(I):  (1/5.C)+(4/5.(1/4.C)-(3/4.0))-S-I=0
    VE(I):  20M-I

    VE(I)-VE(~I)=20M-I
    Aceptariamos cuando el costo del informe fuera menos o igual de 20M
